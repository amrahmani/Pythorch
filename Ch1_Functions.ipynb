```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
  return x**2

# Define a small value for h (step size)
h = 0.1

# Generate x values for plotting
x = np.linspace(-4, 4, 1000)  # Adjust range as needed

# Calculate function values
y = f(x)

# Calculate numerical derivative using forward difference
dy_dx = (f(x + h) - f(x)) / h

# Calculate exact derivative (for comparison)
exact_dy_dx = 2*x  # Derivative of x^2 is 2x

# Find x and y values for the slope at x = 2
x_slope = 2
y_slope = f(x_slope)
dy_dx_slope = (f(x_slope + h) - f(x_slope)) / h  # Numerical derivative at x=2

# Plot the function and derivative
plt.plot(x, y, label='f(x) = x^2')
plt.plot(x, dy_dx, label='Numerical Derivative')
plt.plot(x, exact_dy_dx, label='Exact Derivative (for comparison)')

# Plot the slope line at x=2
plt.axvline(x=x_slope, color='r', linestyle='--', label='x = 2')
plt.axhline(y=y_slope, color='r', linestyle='--')

# Annotate the slope
plt.annotate(f"Slope at x=2: {dy_dx_slope:.2f}", xy=(x_slope, y_slope + 0.2), color='r')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function, Numerical Derivative, and Slope')
plt.legend()
plt.grid(True)
plt.show()
```

**Nested function**

```python
def outer_function(x):
  """
  This is the outer function. It takes a value x as input.
  """
  def inner_function(y):
    """
    This is the inner function. It takes a value y as input
    and is accessible only within the outer_function.
    """
    return x * y**2  # Inner function uses the value of x from outer function

  # Call the inner function with a value
  result = inner_function(2)
  return result + x  # Outer function uses the result from inner function

# Call the outer function
final_result = outer_function(3)
print(final_result)  # Output: 15 (3 + (3 * 2^2))
```
