```python
import torch

# Example 1: Creating a tensor from a list
tensor1 = torch.tensor([1, 2, 3])
print("Tensor 1:", tensor1)

# Example 2: Creating a tensor filled with zeros
tensor2 = torch.zeros(2, 3)
print("Tensor 2:", tensor2)

# Example 3: Creating a tensor filled with ones
tensor3 = torch.ones(3, 2)
print("Tensor 3:", tensor3)

# Example 4: Creating a tensor with random values
tensor4 = torch.rand(2, 2)
print("Tensor 4:", tensor4)

# Example 5: Accessing elements of a tensor
print("Element at (1, 1) in Tensor 4:", tensor4[1, 1])

# Example 6: Reshaping a tensor
tensor5 = torch.arange(1, 10)
reshaped_tensor5 = tensor5.view(3, 3)
print("Reshaped Tensor 5:", reshaped_tensor5)

# Example 7: Adding tensors
tensor6 = torch.tensor([[1, 2], [3, 4]])
tensor7 = torch.tensor([[5, 6], [7, 8]])
sum_tensor = tensor6 + tensor7
print("Sum of Tensor 6 and Tensor 7:", sum_tensor)

# Example 8: Element-wise multiplication
product_tensor = tensor6 * tensor7
print("Element-wise Product of Tensor 6 and Tensor 7:", product_tensor)

# Example 9: Matrix multiplication
matmul_result = torch.matmul(tensor6, tensor7)
print("Matrix Multiplication Result:", matmul_result)

# Example 10: Computing mean of a tensor
mean_tensor = torch.mean(tensor6.float())
print("Mean of Tensor 6:", mean_tensor)

# Example 11: Computing standard deviation of a tensor
std_tensor = torch.std(tensor6.float())
print("Standard Deviation of Tensor 6:", std_tensor)

# Example 12: Computing cosine similarity between two tensors
cos_sim = torch.cosine_similarity(tensor6.float().view(-1), tensor7.float().view(-1), dim=0)
print("Cosine Similarity between Tensor 6 and Tensor 7:", cos_sim)

# Example 13: Creating a tensor from numpy array
import numpy as np
numpy_array = np.array([[1, 2], [3, 4]])
tensor_from_numpy = torch.from_numpy(numpy_array)
print("Tensor from NumPy Array:", tensor_from_numpy)

# Example 14: Moving tensor to GPU
if torch.cuda.is_available():
    tensor6_gpu = tensor6.cuda()
    print("Tensor 6 moved to GPU:", tensor6_gpu)

# Example 15: Checking tensor shape
print("Shape of Tensor 6:", tensor6.shape)

# Example 16: Checking tensor data type
print("Data type of Tensor 6:", tensor6.dtype)

# Example 17: Creating a tensor with specific data type
tensor8 = torch.tensor([1, 2, 3], dtype=torch.float64)
print("Tensor 8 with dtype torch.float64:", tensor8)

# Example 18: Cloning a tensor
cloned_tensor6 = tensor6.clone()
print("Cloned Tensor 6:", cloned_tensor6)

# Example 19: Slicing a tensor
sliced_tensor6 = tensor6[:, 1]
print("Sliced Tensor 6:", sliced_tensor6)

# Example 20: Concatenating tensors
concatenated_tensor = torch.cat((tensor6, tensor7), dim=1)
print("Concatenated Tensor:", concatenated_tensor)

# Example 21: Transposing a tensor
transposed_tensor6 = torch.transpose(tensor6, 0, 1)
print("Transposed Tensor 6:", transposed_tensor6)

# Example 22: Getting maximum element and its index
max_value, max_index = torch.max(tensor6, dim=1)
print("Maximum value and index of Tensor 6:", max_value, max_index)

# Example 23: Getting minimum element and its index
min_value, min_index = torch.min(tensor6, dim=1)
print("Minimum value and index of Tensor 6:", min_value, min_index)

# Example 24: Sorting a tensor
sorted_tensor6, sorted_indices = torch.sort(tensor6, dim=1, descending=True)
print("Sorted Tensor 6:", sorted_tensor6)

# Example 25: Applying element-wise function to a tensor
tensor9 = torch.tensor([-1, -2, 3, 4])
abs_tensor9 = torch.abs(tensor9)
print("Absolute values of Tensor 9:", abs_tensor9)

# Example 26: Applying element-wise function to a tensor with inplace operation
tensor9.abs_()
print("Tensor 9 after applying inplace abs:", tensor9)

# Example 27: Applying softmax function to a tensor
softmax_tensor6 = torch.softmax(tensor6.float(), dim=1)
print("Softmax of Tensor 6:", softmax_tensor6)

# Example 28: Creating a tensor with gradients
tensor11 = torch.tensor([1, 2, 3], dtype=torch.float, requires_grad=True)
print("Tensor 11 with gradients enabled:", tensor11)

# Example 29: Performing backpropagation on a tensor
loss = tensor11.sum()
loss.backward()
print("Gradients of Tensor 11:", tensor11.grad)
```
